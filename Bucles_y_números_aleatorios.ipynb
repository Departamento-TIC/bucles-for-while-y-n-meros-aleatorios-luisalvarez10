#6. Escribe un programa en Python que:
* Le pida al usuario siete números
* Imprima la suma total de esos números
* Imprima la cuenta de las entradas positivas, las que sean iguales a cero, y las negativas. Emplea una cadena if, elif, else, en lugar de tres if consecutivos

````python
n1 = int(input("Introduzca un número entero: "))
n2 = int(input("Introduzca otro número entero: "))
n3 = int(input("Introduzca otro número entero: "))
n4 = int(input("Introduzca otro número entero: "))
n5 = int(input("Introduzca otro número entero: "))
n6 = int(input("Introduzca otro número entero: "))
n7 = int(input("Introduzca otro número entero: "))

numbers = [n1, n2, n3, n4, n5, n6, n7]

print(f"\nLa suma de esos números es {sum(numbers)}\n")

for i in numbers:
  if i > 0:
    print(f"{i} es positivo")
  elif i < 0:
    print(f"{i} es negativo")
  else:
    print(f"{i} es igual a 0")

````

#7. Lanzador de monedas:
* Crea un programa que imprima al azar 0 ó 1.
* En lugar de 0 o 1, que imprima cara o cruz.
* Añádele un bucle para que lo haga 50 veces.
* Crea una suma acumulada para el total de veces que salió cara y las que salió cruz.


````python
import random

face = 0
tails = 0

for i in range(50):
  number = random.randint(0, 1)

  if number == 0:
    print("Cara")
    face += 1
  else:
    print("Cruz")
    tails += 1

print(f"\nHa salido cara {face} veces")
print(f"Ha salido cruz {tails} veces")
````

#8. Escribe un programa para jugar a piedra, tijera y papel:

* Crea un programa que imprima aleatoriamente 0, 1, ó 2.
* Usando sentencias if, expande el programa de manera que ahora imprima al azar piedra, papel o tijera. No hagas la selección desde una lista, tal como hemos visto en el capítulo.
* Añade al programa la opción de que primero le pregunte al usuario qué es lo que elige.(Sería más fácil darle a escoger entre las opciones 1, 2, ó 3.)
* Añade una declaración condicional para determinar quién gana.


````python
print("1. Piedra\n2. Papel\n3. Tijera")
user_answer = int(input("Elija una de las siguientes opciones(Introduzca solo el número):"))

machine_answer = random.randint(1, 3)

print("\n")
if machine_answer == user_answer:
  print("Habéis quedado empate")
elif user_answer == 1 and machine_answer == 3:
  print("La máquina ha sacado tijera")
  print("Has ganado")
elif user_answer == 2 and machine_answer == 1:
  print("La máquina ha sacado piedra")
  print("Has ganado")
elif user_answer == 3 and machine_answer == 2:
  print("La máquina ha sacado papel")
  print("Has ganado")
else:
  if machine_answer == 1:
    print("La máquina ha sacado piedra")
  elif machine_answer == 2:
    print("La máquina ha sacado papel")
  else:
    print("La máquina ha sacado tijera")
  print("Has perdido")
````
